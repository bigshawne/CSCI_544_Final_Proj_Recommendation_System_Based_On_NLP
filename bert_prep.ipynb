```python
start_time = time.time()

res_dict = {}
user_review_dict = {}
saved_user_review_dict = {}
train_r_id_list = list(test_data.keys())
#train_r_id_list = train_r_id_list[:1000]
for each_review in train_r_id_list:
  #each_info = review_info_data.get(each_review)
  #u_id = each_info.get('user_id')
  #bus_id = each_info.get('business_id')
  #current_user_set = res_dict.get(bus_id, [])
  #current_user_set.append(u_id)
  #res_dict[bus_id] = current_user_set

  each_text = review_text_data.get(each_review)
  bert_encoding = model.encode([each_text])
  #current_res_set = user_review_dict.get(u_id, {})
  #current_res_set[bus_id] = bert_encoding
  #user_review_dict[u_id] = current_res_set

  # save user_review_dict : list(bert_encoding[0])
  encoding_list = list(bert_encoding[0])
  encoding_list = [float(i) for i in encoding_list]
  #temp_set = saved_user_review_dict.get(each_review, {})
  #temp_set[bus_id] = encoding_list
  saved_user_review_dict[each_review] = encoding_list

print("--- %s seconds ---" % (time.time() - start_time))
```
